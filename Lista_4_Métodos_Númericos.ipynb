# - Dosagem de Medicamentos:

Inicialmente podemos apresentar uma distinção entre os termos Dose e Dosagem. A Dose está relacionada com a quantidade de fármaco em um medicamento, ou seja, é a quantidade de substância que pode ter impacto num organismo, substância essa responsável pelo efeito terapêutico. Já a Dosagem pode ser definida como o ato de dosar ou de medir a quantidade de dose que será administrada a um organismo em um determinado intervalo de tempo, podendo ser necessário a realização de cálculos para determinar a dosagem correta de algum medicamento.

Além disso, a dose pode ser dividida em seis categorias diferentes que delimitam o seu objetivo principal:

*   **Dose de Ataque**: realizada para elevar rapidamente a concentração do fármaco na corrente sanguínea do organismo, obtendo um rápido resultado terapêutico;
*   **Dose de Manutenção**: administrada em intervalos regulares de tempo, utilizada para manter uma certa concentração de fármaco na quantidade desejada;
*   **Dose Mínima**: menor dose eficaz que possui resultado terapêutico;
*   **Dose Máxima**: maior dose eficaz onde não são observados efeitos tóxicos;
*   **Dose Tóxica**: dose maior que a eficaz, possuindo efeitos tóxicos, também chamada de "overdose";
*   **Dose Letal**: quantidade que pode levar o organismo a falecer.

Nesse sentido, é preciso tomar atenção para o desafio da dosagem correta de medicamentos, sendo necessário manter a concentração do mesmo entre a Dose Mínima e a Dose Máxima. Assim, é importante salientar que cálculos incorretos ou inadequados de dosagens podem gerar ineficiência, ou superdosagem (intoxicação), afetando negativamente a saúde do organismo. Devido a isso, finalmente, evidencia-se que saber a dosagem correta de um medicamento, assim como o intervalo de tempo entre cada dose, pode ser um problema para o médico que irá prescrever o medicamento.

Dessa forma, uma solução para o problema evidenciado acima será descrita ao longo do relatório.

# -  Proposta


Chamando o limite superior de concentração da droga no sangue de $H$ e o limite inferior de $L$, desejamos que seja prescrita uma dosagem $C0$ com um intervalo de tempo $T$ entre cada dose, de forma a manter a concentração entre $H$ e $L$.

O decaimento de concentração da droga com o tempo pode ser modelado pela equação:

\begin{equation}
dC/dt = -kC
\end{equation}

Onde $k$ é uma constante denominada constante de eliminação da droga. O valor inicial da concentração é $C0 = H$ e a próxima dose ocorrerá quando a concentração for inferior a $L$, após $T$ horas.

Dessa forma, a proposta é utilizar o Método de Euler para obter uma equação de diferenças finitas da equação diferencial de decaimento e, em seguida, calcular o decaimento da concentração de um medicamento ao longo do tempo e o tempo $T$ necessário entre cada dosagem. Será feito também um gráfico da concentração em função do tempo, além da análise dos resultados obtidos.

**`- Bibliotecas e variáveis que vão ser utilizadas ao longo da atividade:`**

```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
import pandas as pd


K = 0.44
L = 0.5
H = 5

T = np.arange(0, 12.1, 0.1)
```

# - Solução Pelo Método de Euler

  Em matemática e ciência computacional, o método de Euler, cujo nome relaciona-se com Leonhard Euler, é um procedimento numérico de primeira ordem para solucionar equações diferenciais ordinárias com um valor inicial dado. É o tipo mais básico de método explícito para integração numérica para equações diferenciais ordinárias:

\begin{equation}
Yn+1= Yn+hf(tn,yn)
\end{equation}

$Yn+1$ = valor que queremos encontrar de concentração $C$.

$Yn$ = valor inicial.

$h$ = passo.

$f(tn,yn)$ = é a nossa função.


- Dessa forma: 

 - Escolhemos o valor $0.1$ para $h$ que é o tamanho de cada passo (intervalo) do eixo $x$. Que representa de quantas em quantas horas nosso tempo aumenta.
 - Nosso $Yn = C(n)$ e inicialmente $= H$ (limite superior de concentração da droga no sangue).

 - Nossa função fica sendo: $ -kC =  -kYn  =  -kC(n)$

- Assim o método fica igual a:

\begin{equation}
C = C(n)+h(-KC(n))
\end{equation}

Agora basta aplicarmos aos códigos para resolvermos as concentracões em todos os intervalos determinados. Sempre com a atualização das variáveis que nessecitam disto.




```python
#Vetor para armazenar a concentraçao do remedio a cada intervalo de hora. E inicializando com a variável H(limite superior), que é nossa dose máxima:
C = np.zeros(121)
C[0] = H

#Definição do passo h, cosiderando 120 subintervalos de tempo:
h = 0.1
n = 120

#Definição da EDO:
def f(t, c):
  w = -K*c
  return w

#Calculando a concentração em função do tempo:
for i in range (n):
  C[i+1] = C[i] + h * f(T[i], C[i])
```

```python
#Achando o Tempo Necessário Entre Cada Dosagem:
for i in range(n):
  if C[i] < L:
    te = T[i]
    print("- O tempo entre cada dosagem é:", te, "horas, ou", te*60, "minutos.")
    break;
```

**`- Gráfico para uma melhor visualização e analise dos resultados:`**

```python
#coordenadas do limite inferior
LI = np.zeros(121)
for i in range (len(LI)):
  LI[i] = L


plt.rcParams["figure.figsize"] = (22, 11)
plt.xticks(range(0, 13, 1))
plt.plot(T, C, 'b-o')
plt.plot(T, LI, 'r-')
plt.grid(True)
plt.xlabel('Tempo (h)')
plt.ylabel('Concentração(mg/mL)')
plt.title('Decaimento da concentração de um medicamento no sangue ao longo do tempo')
plt.legend(['Concentração', 'Limite inferior'])

plt.show;
```

#Resolvendo Pelo Método Analítico


Pelo método analítico devemos resolver a equação do decaimento de concentração, achando:

\begin{equation}
C = e^{-Kt+G}
\end{equation}

Devemos achar a constante G, substituindo k = 0.44, t = 0 e C = H = 5, (esses valores foram disponibilizados para nós, para essas respectivas variáveis).

Assim, achamos G = ln(5), aproximadamente 1.60943.

Dessa forma, ficamos com a equação:

\begin{equation}
C = e^{-Kt+1.60943}
\end{equation}

Portanto, basta resolve-lá para acharmos as concentrações no intervalo de tempo escolhido.

```python
#Vetor para armazenar a concentraçao do remedio
C_anali = np.zeros(121)
C_anali[0]=H

#Definição da Concentração Baseado no Tempo:
for i in range(n):
  C_anali[i+1]= math.e**((-K*T[i+1])+1.60943)
```

```python
#Achando o tempo necessário entre cada dosagem:
for i in range(n):
  if C_anali[i] < L:
    ta = T[i]
    print("O tempo entre cada dosagem é:", round(ta, 1 ), "horas, ou", round(ta*60, 1), "minutos.")
    break;
```

**`- Gráfico para uma melhor visualização e analise dos resultados:`**

```python
plt.rcParams["figure.figsize"] = (22, 11)
plt.xticks(range(0, 13, 1))
plt.plot(T, C_anali, 'b-o')
plt.plot(T, LI, 'r-')
plt.grid(True)
plt.xlabel('Tempo (h)')
plt.ylabel('Concentração(mg/mL)')
plt.title('Decaimento da concentração de um medicamento no sangue ao longo do tempo')
plt.legend(['Concentração', 'Limite inferior'])

plt.show;
```
